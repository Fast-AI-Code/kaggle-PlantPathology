# Data

```python
path = Path("data/")
```

## Label process

```python
labels = pd.read_csv('data/train.csv')
```

```python
labels.head()
```

```python
labels['healthy'].replace(1,'healthy')
```

```python
for a in labels.columns[1::]:
    labels[a] = labels[a].replace(1,a[0])
    labels[a] = labels[a].replace(0,'n'+a[0])
```

```python
labels['join'] = labels.iloc[:,1::].astype('str').agg(' '.join, axis=1)
```

```python
labels[['image_id','join']].to_csv('data/temp.csv')
```

## Load data

```python
data = (
    (
        ImageList.from_csv(path, "temp.csv", folder="images/Train", suffix=".jpg")
        .split_by_rand_pct()
        .label_from_df(label_delim=' ',cols = ['join'])
        .transform(get_transforms(), size=128)
    )
    .databunch(bs=64)
    .normalize(imagenet_stats)
)
```

```python
data.show_batch(4)
```

```python
data.c
```

```python
data
```

```python
learn = None
gc.collect()
```

# Models

## Resnet34

```python
learn = cnn_learner(
    data, models.resnet34, metrics=[accuracy_thresh], opt_func=AdamW, callback_fns=ShowGraph
).to_fp16()
```

```python
learn.lr_find()
learn.recorder.plot()
```

```python
learn.fit_one_cycle(20, 2e-3)
```

```python
learn.show_results()
```

```python
learn.unfreeze()
learn.fit_one_cycle(20, 2e-3)
```

```python
learn.validate()
```

```python
learn.show_results()
```
